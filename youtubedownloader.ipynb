```python
# Program to download youtube videos
# download the pytube package
#!pip install pytube
# Importing the module
import pytube 

video_default = 'https://www.youtube.com/watch?v=99k-EAMBuM8'
# Taking input the link
link = video_default # input('Paste the url here : ')

# Using the module to download the video
video = pytube.YouTube(link)
video.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first().download()

#promt to make sure that the download is complete
print("Downloaded Successfully, ",link)
print(f'title = {video.title}', f', video views = {video.views}', f', length = {video.length} seconds')
```
